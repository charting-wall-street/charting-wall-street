```python
import matplotlib.pyplot as plt

double_top = { 3: 126480, 17: 8526, 29: 3551, 5: 58100, 7: 35049, 11: 17486, 13: 13326, 15: 10581, 23: 5086, 25: 4407, 9: 23964, 27: 3880, 19: 7098, 21: 5996, 31: 3184 }

x = list(double_top.keys())
y = list(double_top.values())

plt.bar(x, y)
plt.xlabel('Range Parameter (High-Low)')
plt.ylabel('Occurrences')
plt.title('Double-Top Occurrences for a Given High/Low range')
plt.savefig(f"./figures/occurrences_double-top_range.svg", format="svg")
plt.show()
```

```python
double_bottom = { 31: 2610, 3: 115809, 7: 31894, 21: 5081, 5: 53483, 15: 8965, 25: 3698, 9: 21321, 11: 15392, 27: 3258, 23: 4263, 29: 2893, 13: 11559, 17: 7246, 19: 6015 }

x = list(double_bottom.keys())
y = list(double_bottom.values())

plt.bar(x, y)
plt.xlabel('Range Parameter (High-Low)')
plt.ylabel('Occurrences')
plt.title('Double-Bottom Occurrences for a Given High/Low range')
plt.savefig(f"./figures/occurrences_double-bottom_range.svg", format="svg")
plt.show()
```

```python
import numpy as np

# Extract keys and values for double-top and double-bottom occurrences
top_x = sorted(double_top.keys())
top_y = [double_top[key] for key in top_x]

bottom_x = sorted(double_bottom.keys())
bottom_y = [double_bottom[key] for key in bottom_x]

# Set the width for each bar
bar_width = 0.35

# Calculate the center positions of the bars
top_x_pos = np.arange(len(top_x))
bottom_x_pos = [x + bar_width for x in top_x_pos]

# Plot the bars for double-top and double-bottom occurrences
plt.bar(top_x_pos, top_y, width=bar_width, align='center', label='Double-Top')
plt.bar(bottom_x_pos, bottom_y, width=bar_width, align='center', label='Double-Bottom')

# Set the x-axis ticks and labels
plt.xticks(top_x_pos + bar_width/2, top_x)
plt.xlabel('Range Parameter (High-Low)')
plt.ylabel('Occurrences')
plt.title('Double-Top and Double-Bottom Occurrences for a Given High/Low Range')
plt.savefig(f"./figures/occurrences_double-bottom_double-top_range.svg", format="svg")
plt.legend()

plt.show()
```
