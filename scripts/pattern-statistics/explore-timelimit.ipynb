```python
import pandas as pd
csv_dir = "~/Projects/strategy-evaluator/output/csv"
```

```python
import matplotlib.pyplot as plt

def plot_occurrences(algorithm, evaluator):
    df = pd.read_csv(f"{csv_dir}/size/by-limit_{evaluator}_{algorithm}.csv", delimiter=',', index_col=0)
    df_flattened = df.sum()

    ax = df_flattened.plot(kind='bar')
    plt.xlabel('Limit')
    plt.ylabel('Number of Occurrences')

    # Get the current tick labels
    xlabels = [label.get_text() for label in ax.get_xticklabels()]

    # Format the labels
    if xlabels[0][:len("limit")] == "limit":
        ax.set_xlabel("Limit")
        xlabels = [label.replace("limit:", "") for label in xlabels]
    else:
        ax.set_xlabel("Range")
        xlabels = [label.replace("rng:", "") for label in xlabels]

    ax.set_xticklabels(xlabels)
        
    plt.title(f"Occurrences/Limit for {algorithm} ({evaluator})".title())
    plt.savefig(f"./figures/occurrences_{evaluator}_{algorithm}.svg", format="svg")
    plt.show()
```

```python
plot_occurrences("double-bottom", "buckets")
```

```python
plot_occurrences("double-bottom", "barriers")
```

```python
plot_occurrences("random", "barriers")
```
