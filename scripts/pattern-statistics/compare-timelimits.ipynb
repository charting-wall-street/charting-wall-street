```python
import pandas as pd

df_db_limit7 = pd.read_csv("./comparisons/timelimits/by-range_barriers_double-bottom_limit7.csv", delimiter=',', index_col=0)
df_db_limit14 = pd.read_csv("./comparisons/timelimits/by-range_barriers_double-bottom_limit14.csv", delimiter=',', index_col=0)
df_db_limit21 = pd.read_csv("./comparisons/timelimits/by-range_barriers_double-bottom_limit21.csv", delimiter=',', index_col=0)
```

```python
table_7_vs_14 = df_db_limit14 - df_db_limit7
```

```python
table_14_vs_21 = df_db_limit21 - df_db_limit14
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from matplotlib.colors import LinearSegmentedColormap
colors = ["red", "white", "green"]
cmap = LinearSegmentedColormap.from_list("custom", colors, N=100)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(table_7_vs_14, annot=True, cmap=cmap, center=0, vmin=-10, vmax=10)
plt.title("Double Bottom Difference: Time Limit 7 vs Time Limit 14 (Triple Barrier)")
plt.tight_layout()
plt.savefig("./figures/timelimit_7_vs_14.svg", format="svg")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(table_14_vs_21, annot=True, cmap=cmap, center=0, vmin=-10, vmax=10)
plt.title("Double Bottom Difference: Time Limit 14 vs Time Limit 21 (Triple Barrier)")
plt.tight_layout()
plt.savefig("./figures/timelimit_14_vs_21.svg", format="svg")
plt.show()
```
