```python
import pandas as pd

df_db_barriers = pd.read_csv("./comparisons/evaluator/by-range_barriers_double-bottom.csv", delimiter=',', index_col=0)
df_db_buckets = pd.read_csv("./comparisons/evaluator/by-range_buckets_double-bottom.csv", delimiter=',', index_col=0)
df_dt_barriers = pd.read_csv("./comparisons/evaluator/by-range_barriers_double-top.csv", delimiter=',', index_col=0)
df_dt_buckets = pd.read_csv("./comparisons/evaluator/by-range_buckets_double-top.csv", delimiter=',', index_col=0)
```

```python
table_double_bottom = df_db_barriers - df_db_buckets
```

```python
table_double_top = df_dt_barriers - df_dt_buckets
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from matplotlib.colors import LinearSegmentedColormap
colors = ["green", "white", "red"]
cmap = LinearSegmentedColormap.from_list("custom", colors, N=100)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(table_double_bottom, annot=True, cmap=cmap, center=0, vmin=-10, vmax=10)
plt.title("Double Bottom Difference: Triple Barrier vs Hybrid Bucket-based")
plt.tight_layout()
plt.savefig("./figures/eval_comp_double_bottom.svg", format="svg")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(table_double_top, annot=True, cmap=cmap, center=0, vmin=-10, vmax=10)
plt.title("Double Top Difference: Triple Barrier vs Hybrid Bucket-based")
plt.tight_layout()
plt.savefig("./figures/eval_comp_double_top.svg", format="svg")
plt.show()
```
